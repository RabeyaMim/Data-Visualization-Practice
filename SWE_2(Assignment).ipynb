**1.Flow graph visualization**

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
g = nx.DiGraph()

# Add nodes for each step
g.add_node("1")
g.add_node("2")
g.add_node("3")
g.add_node("4")
g.add_node("5")
g.add_node("6")
g.add_node("7")
g.add_node("8")
g.add_node("9")
g.add_node("10")
g.add_node("11")
g.add_node("12")
g.add_node("13")

# Add edges between steps
g.add_edge("1", "2")
g.add_edge("2", "3")
g.add_edge("3", "4")
g.add_edge("4", "5")
g.add_edge("5", "6")
g.add_edge("6", "8")
g.add_edge("8", "9")
g.add_edge("9", "2")

g.add_edge("1", "2")
g.add_edge("2", "10")
g.add_edge("10", "11")
g.add_edge("11", "13")

g.add_edge("1", "2")
g.add_edge("2", "10")
g.add_edge("10", "12")
g.add_edge("12", "13")

g.add_edge("1", "2")
g.add_edge("2", "3")
g.add_edge("3", "10")
g.add_edge("10", "11")
g.add_edge("11", "13")

g.add_edge("1", "2")
g.add_edge("2", "3")
g.add_edge("3", "4")
g.add_edge("4", "5")
g.add_edge("5", "6")
g.add_edge("6", "7")
g.add_edge("7", "8")
g.add_edge("8", "9")
g.add_edge("9", "2")

g.add_edge("1", "2")
g.add_edge("2", "3")
g.add_edge("3", "4")
g.add_edge("4", "5")
g.add_edge("5", "8")
g.add_edge("8", "9")
g.add_edge("9", "2")



# Draw the graph
pos = nx.spring_layout(g)
nx.draw(g, pos, with_labels=True, node_size=600, font_size=15)
plt.show()
```

**2.Cyclomatic complexity of the flow graph**

```python
# Calculate the cyclomatic complexity
E = g.number_of_edges()
N = g.number_of_nodes()
Vg = E - N + 2

print("Cyclomatic Complexity:", Vg)


```

**3.Linearly independent paths**

```python


# Determine basis set of linearly independent paths
basis_set = []

# Path 1: 1 -> 2 -> 10 -> 11 -> 13
basis_set.append(["1", "2", "10", "11", "13"])

# Path 2: 1 -> 2 -> 10 -> 12 -> 13
basis_set.append(["1", "2", "10", "12", "13"])

# Path 3: 1 -> 2 -> 3 -> 10 -> 11 -> 13
basis_set.append(["1", "2", "3", "10", "11", "13"])

# Path 4: 1 -> 2 -> 3 -> 4 -> 5 -> 8 -> 9 -> 2
basis_set.append(["1", "2", "3", "4", "5", "8", "9", "2"])

# Path 5: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 8 -> 9 -> 2
basis_set.append(["1", "2", "3", "4", "5", "6", "8", "9", "2"])

# Path 6: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8 -> 9 -> 2
basis_set.append(["1", "2", "3", "4", "5", "6", "7", "8", "9", "2"])

# Draw the graph
pos = nx.spring_layout(g)
nx.draw(g, pos, with_labels=True, node_size=500, font_size=10)
plt.show()

# Print the basis set of linearly independent paths
for i, path in enumerate(basis_set):
    print(f"Path {i+1}: {' -> '.join(path)}")
```

```python
class AverageProcedure:
    @staticmethod
    def average_calculation(values, minimum, maximum):
        i = 0
        total_input = 0
        total_valid = 0
        _sum = 0

        while i < len(values) and values[i] != -999 and total_input < 100:
            total_input += 1
            if minimum <= values[i] <= maximum:
                total_valid += 1
                _sum += values[i]
            i += 1

        if total_valid > 0:
            return float(_sum) / total_valid
        else:
            return -999


# Test Cases
# Test Case 1: Path 1 - 1-2-10-11-13
values1 = [5, 10, -999]
minimum1 = 0
maximum1 = 20
expected_result1 = 7.5

result1 = AverageProcedure.average_calculation(values1, minimum1, maximum1)
print(f"Test Case 1 - Expected Result: {expected_result1}, Actual Result: {result1}")

# Test Case 2: Path 2 - 1-2-10-12-13
values2 = [5, 25, -999]
minimum2 = 0
maximum2 = 20
expected_result2 = 4

result2 = AverageProcedure.average_calculation(values2, minimum2, maximum2)
print(f"Test Case 2 - Expected Result: {expected_result2}, Actual Result: {result2}")

# Test Case 3: Path 3 - 1-2-3-10-11-13
values3 = [5, 10, 15, -999]
minimum3 = 0
maximum3 = 20
expected_result3 = 10
result3 = AverageProcedure.average_calculation(values3, minimum3, maximum3)
print(f"Test Case 3 - Expected Result: {expected_result3}, Actual Result: {result3}")


# Test Case 4: 1 -> 2 -> 3 -> 4 -> 5 -> 8 -> 9 -> 2
values4 = [2, 10, 25, -999]
minimum4 = 0
maximum4 = 20
expected_result4 = 6
result4 = AverageProcedure.average_calculation(values4, minimum4, maximum4)
print(f"Test Case 4 - Expected Result: {expected_result4}, Actual Result: {result4}")



# Test Case 5: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 8 -> 9 -> 2
values5 = [5, 1, 15, -999]
minimum5 = 0
maximum5 = 20
expected_result5 = 12
result5 = AverageProcedure.average_calculation(values5, minimum5, maximum5)
print(f"Test Case 5 - Expected Result: {expected_result5}, Actual Result: {result5}")


# Test Case 6: 1 -> 2 -> 3 -> 4 -> 5 -> 6 -> 7 -> 8 -> 9 -> 2
values6 = [9, 10, 15]
minimum6 = 0
maximum6 = 20
expected_result_6 = 11.33
result6 = AverageProcedure.average_calculation(values6, minimum6, maximum6)
print(f"Test Case 6 - Expected Result: {expected_result_6}, Actual Result: {result6}")
```
